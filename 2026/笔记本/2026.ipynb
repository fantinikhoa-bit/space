```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 数据准备：基于美国/商业航天近年来的爆发式增长趋势
# 2025年的165次符合从2020年(44次) -> 2023年(116次)的增长逻辑
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2025])
launches = np.array([15, 18, 13, 19, 23, 20, 22, 29, 31, 21, 44, 51, 78, 116, 165])

# 定义 Logistic 模型
def logistic_model(t, K, A, r):
    t_norm = t - 2010
    return K / (1 + A * np.exp(-r * t_norm))

# 拟合模型
# 放宽 K (上限) 的边界以寻找自然饱和点
p0 = [300, 50, 0.2]
bounds = ([165, 0, 0], [10000, 1000, 2.0])

try:
    popt, pcov = curve_fit(logistic_model, years, launches, p0=p0, bounds=bounds, maxfev=10000)
    K_fit, A_fit, r_fit = popt
except Exception as e:
    K_fit, A_fit, r_fit = 0, 0, 0

# 输出结果
if K_fit > 0:
    val_2050 = logistic_model(2050, K_fit, A_fit, r_fit)
    print(f"年发射上限 (K): {K_fit:.2f}")
    print(f"2050年预测值: {val_2050:.2f}")
    
    # 绘图（已保存到桌面）
    future_years = np.arange(2010, 2051)
    predicted_launches = logistic_model(future_years, K_fit, A_fit, r_fit)
    plt.figure(figsize=(10, 6))
    plt.scatter(years, launches, color='red', label='Historical Data (incl. 2025=165)')
    plt.plot(future_years, predicted_launches, color='blue', label=f'Logistic Fit (Limit={K_fit:.0f})')
    plt.title('Logistic Prediction of Annual Rocket Launches')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.savefig('rocket_launch_prediction.png')
```
